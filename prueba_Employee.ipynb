```python
def plot_int_distributions(dataframe, columns_per_row=3, color='steelblue'):
    """
    Genera histogramas con KDE para visualizar la distribuci√≥n de columnas enteras.

    Par√°metros:
        dataframe (pd.DataFrame): DataFrame con los datos a visualizar.
        columns_per_row (int): N√∫mero de gr√°ficos por fila. Default: 3.
        color (str): Color de los gr√°ficos. Default: 'lightseagreen'.
    """
    # Filtrar solo columnas de tipo entero
    data_int = dataframe.select_dtypes(include=['int64'])

    if data_int.empty:
        print("No hay columnas de tipo entero en el DataFrame.")
        return

    # Calcular el n√∫mero de filas din√°micamente
    num_cols = len(data_int.columns)
    rows = -(-num_cols // columns_per_row)  # Divisi√≥n redondeada hacia arriba

    # Configurar la figura
    fig, axes = plt.subplots(rows, columns_per_row, figsize=(5 * columns_per_row, 3 * rows))
    axes = axes.flatten()  # Convertir en lista para facilitar el acceso

    # Graficar distribuciones con histplot (reemplazo de distplot)
    for idx, col in enumerate(data_int.columns):
        sns.histplot(data=data_int[col], kde=True, ax=axes[idx], color=color)
        axes[idx].set_title(f"Distribuci√≥n de {col}", fontsize=12, fontweight='bold')
        axes[idx].set_xlabel("Valor", fontsize=10)
        axes[idx].set_ylabel("Frecuencia", fontsize=10)
        axes[idx].grid(True, linestyle='--', alpha=0.5)

    # Ocultar ejes vac√≠os si hay menos columnas que espacios en la cuadr√≠cula
    for i in range(idx + 1, len(axes)):
        fig.delaxes(axes[i])

    plt.tight_layout()
    plt.show()

# Llamar a la funci√≥n con el DataFrame
plot_int_distributions(data_num)
```

```python
# Funci√≥n para eliminar caracter√≠sticas altamente correlacionadas
def remove_correlated_features(data, threshold=0.75):
    correlation_matrix = data.corr()
    correlated_features = set()

    for col in correlation_matrix.columns:
        for row in correlation_matrix.index:
            if col != row and abs(correlation_matrix.loc[col, row]) > threshold:
                correlated_features.add(col)

    return correlated_features
```

```python
# Identificar y eliminar caracter√≠sticas con correlaci√≥n mayor a 0.75
correlated_features = remove_correlated_features(numerical, 0.75)
numerical.drop(columns=correlated_features, inplace=True)
```

## Seleccion de Modelos

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
# Funci√≥n para entrenar y evaluar el modelo
def model_eval(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)

    # Predicciones en entrenamiento y prueba
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Evaluaci√≥n en conjunto de entrenamiento
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_precision = precision_score(y_train, y_train_pred)
    train_recall = recall_score(y_train, y_train_pred)

    print("Evaluaci√≥n en Entrenamiento:")
    print(f'Accuracy: {train_accuracy:.2f}, Precision: {train_precision:.2f}, Recall: {train_recall:.2f}')
    print("=" * 40)

    # Evaluaci√≥n en conjunto de prueba
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred)
    test_recall = recall_score(y_test, y_test_pred)
    test_f1 = f1_score(y_test, y_test_pred)

    print("Evaluaci√≥n en Prueba:")
    print(f'Accuracy: {test_accuracy:.2f}, Precision: {test_precision:.2f}, Recall: {test_recall:.2f}, F1 Score: {test_f1:.2f}')
    print("=" * 40)

    # Matriz de confusi√≥n
    ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap='Blues')

    # Reporte de clasificaci√≥n
    print("\nReporte de Clasificaci√≥n:")
    print(classification_report(y_test, y_test_pred))
```
